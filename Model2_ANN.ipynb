```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'pH'
#This is for pH model
datapH = pd.read_excel('Polymer_data.xlsx', sheet_name=0)

varpH = ['pH','PAM','HMTA','HQ','T']
X = datapH[varpH]
y = datapH['lnt, lab']
X2 = X[:78]
y2 = y[:78]

XX = X[:78]
yy = y[:78]
print(X)
print(y)
```

```python
#Visualizing gelation time in natural logarithm function
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,1,1)
sns.distplot(datapH['lnt, lab'])
fig.add_subplot(2,1,2)
sns.boxplot(datapH['lnt, lab'])
plt.tight_layout()
```

```python
#visualizing ln(t) of pH, [PAM], [HQ], [HMTA]
fig = plt.figure(figsize=(16,5))
fig.add_subplot(2,2,1)
sns.scatterplot(datapH['pH'], datapH['lnt, lab'])
fig.add_subplot(2,2,2)
sns.scatterplot(datapH['PAM'],datapH['lnt, lab'])
fig.add_subplot(2,2,3)
sns.scatterplot(datapH['HQ'],datapH['lnt, lab'])
fig.add_subplot(2,2,4)
sns.scatterplot(datapH['HMTA'],datapH['lnt, lab'])
```

```python
#visualizing ln(t) of pH, [PAM], [HQ], [HMTA]
fig = plt.figure(figsize=(15,7))
fig.add_subplot(2,2,1)
sns.countplot(datapH['pH'])
fig.add_subplot(2,2,2)
sns.countplot(datapH['PAM'])
fig.add_subplot(2,2,3)
sns.countplot(datapH['HQ'])
fig.add_subplot(2,2,4)
sns.countplot(datapH['T'])
plt.tight_layout()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =0)
```

```python
#standardization scaler - fit&transform on train, fit only on test
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train.astype(np.float))
X_test = s_scaler.transform(X_test.astype(np.float))
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

regr = MLPRegressor(hidden_layer_sizes=(5,100,25), activation='identity', solver='adam', max_iter=2000)
model = regr.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
yy_test = np.array(y_test)
print(regr.score(X_test, y_test), 'R-squared score') #Check in the xls excel to see the R-squared reported in the paper
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, color = 'red')
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'b--', lw=1)
ax.set_title('Gelation Time as Concentration and Temperature Function')
ax.set_xlabel('Measured [hours]')
ax.set_ylabel('Predicted [hours]')
plt.show()
```

```python
# visualizing residuals
fig = plt.figure(figsize=(10,5))
residuals = (y_test- y_pred)
sns.distplot(residuals)
```

```python
#compare actual output values with predicted values
y_pred = regr.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(10)
df1
# evaluate the performance of the algorithm (MAE - MSE - RMSE)
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

```python
Xscaler = s_scaler.fit_transform(X.astype(np.float))
predMPL = regr.predict(Xscaler)

df = pd.DataFrame(predMPL)
df.to_csv('predMPL.csv', index=False)
```
